```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
as2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
as2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
as2020_df
```

```python
# Display the 2021 sales DataFrame
as2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
as2020_df.dtypes
```

```python
# Check the 2021 sales data types.
as2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
sales2020_2021_df = pd.concat([as2020_df, as2021_df]).reset_index(drop=True)
```

```python
# Check if any values are null.
sales2020_2021_df.isnull().sum()
```

```python
# Check the data type of each column
sales2020_2021_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
sales2020_2021_df['invoice_date'] = pd.to_datetime(sales2020_2021_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
sales2020_2021_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_data = sales2020_2021_df.groupby(['region', 'state', 'city'])['units_sold'].sum().rename('Total_Products_Sold').reset_index()

# Show the top 5 results.
grouped_data.head()
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
# Create a pivot table to show the number of products sold for region, state, and city
pivot_table = sales2020_2021_df.pivot_table(values='units_sold', index=['region', 'state', 'city'], aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results
pivot_table.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_data = sales2020_2021_df.groupby(['region', 'state', 'city'])['Total Sales'].sum()

# Show the top 5 results.
grouped_data.to_frame().head().sort_values(by='Total Sales', ascending=False)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_data = sales2020_2021_df.pivot_table(values='Total Sales', index=['region', 'state', 'city'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
pivot_data.head().sort_values(by='Total Sales', ascending=False)
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_data2 = sales2020_2021_df.groupby(['retailer', 'region', 'state', 'city'])['Total Sales'].sum()

# Show the top 5 results.
grouped_data2.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_data2 = sales2020_2021_df.pivot_table(values='Total Sales', index=['retailer', 'region', 'state', 'city'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the top 5 results.
pivot_data2.head().sort_values(by='Total Sales', ascending=False)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.

womens_footwear = sales2020_2021_df.loc[sales2020_2021_df['product'] == 'Women\'s Athletic Footwear']
womens_footwear.head(25)
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
grouped_data3 = womens_footwear.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().rename('Womens_Footwear_Units_Sold').reset_index()

# Show the top 5 results.
grouped_data3.head().sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_data3 = womens_footwear.pivot_table(values='units_sold', index=['retailer', 'region', 'state', 'city'], aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_data3.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'}, inplace=True)
# Show the top 5 results.
pivot_data3.head().sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_data4 = sales2020_2021_df.pivot_table(values='Total Sales', index=['invoice_date'], aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
pivot_data4
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
pivot_data4.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_data4.resample('D').sum().sort_values(by='Total Sales', ascending=True)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
pivot_data4.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_data4.resample('W').sum().sort_values(by='Total Sales', ascending=True)
```
